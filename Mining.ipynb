```python
#traitment of one cell
xp,yp,zp,np=[],[],[],[]
with open(folder+'cell.csv',newline='') as file:
    reader = csv.reader(file)
    for line in reader:
        xp.append(line[1])
        yp.append(line[2])
        zp.append(line[4])
        np.append(line[5])
xp=[float(i) for i in xp]
yp=[float(i) for i in yp]
zpT=[int(i) for i in zp]
np=[int(i) for i in np]
```

```python
cutting_time=60*24
zp=[i%cutting_time for i in zpT]
```

```python
fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xp,yp,zp,c=np,linewidths=0)#
```

```python
def lissage(vector, scheme):
    out=[0]*(len(vector)+len(scheme)-1)
    for add,coef in enumerate(scheme):
        for i,v in enumerate(vector):
            out[i+coef-1]+=v*coef
    return out
    
```
